# El mono de Dawkins

```python
import random
```

```python
#Acá ponemos la frase que vamos a buscar
objetivo="vago no soy,...quiza algo timido para el esjuerzo"
```

```python
#Definimos el conjunto de caracteres que vamos a permitir en la frase
caracteresTodos="abcdefghijklmnopqrstuvwxyz,. " #todas las minúsculas, más punto, coma y espacio
caracteresObj=list(set(objetivo)) #solamente las letras presentes en el objetivo
```

```python
def prueba(frase,caracteres,variantes=100):
    '''
    Función para generar frases "mutadas"
    Toma una frase y devuelve una lista de frases
    donde cada frase cambia en un caracter al azar
    '''
    largoFrase=len(frase)
    largoCaracteres=len(caracteres)
    salida=[]
    for i in range(variantes):
        aMutar=random.randint(0,largoFrase-1)
        nuevoCaracter=caracteres[random.randint(0,largoCaracteres-1)]
        mutada=frase[:aMutar]+nuevoCaracter+frase[aMutar+1:]
#         print (nuevoCaracter)
        salida.append(mutada)
    return salida
        
```

```python
def compara(listaDeFrases, objetivo):
    '''
    Función para comparar una lista de frases con el objetivo
    Devuelve la frase de la lista que es más parecida al objetivo
    '''
    puntajes=[]
    for frase in listaDeFrases:
        iguales=[1 for x,y in zip(objetivo,frase) if x==y]
        puntajes.append(len(iguales))
    return sorted(list(zip(puntajes,listaDeFrases)))[-1][1]
```

```python
#Genero al azar una frase para iniciar la búsqueda
inicial="".join([caracteresTodos[random.randint(0,len(caracteresTodos)-1)] for x in range(len(objetivo))])
```

```python
#Itero la generación de frases mutadas y la comparación
#hasta lograr la frase objetivo
lista=prueba(inicial,caracteresTodos)
generacion=1
while compara(lista,objetivo)!=objetivo:
    laMejor=compara(lista,objetivo)
    lista=prueba(laMejor,objetivo)
    print("generacion %3i: %s"%(generacion,compara(lista,objetivo)))
    generacion+=1
    
```

```python
#Itero la generación de frases mutadas y la comparación
#hasta lograr la frase objetivo
lista=prueba(inicial,caracteresObj)
generacion=1
while compara(lista,objetivo)!=objetivo:
    laMejor=compara(lista,objetivo)
    lista=prueba(laMejor,objetivo)
    print("generacion %3i: %s"%(generacion,compara(lista,objetivo)))
    generacion+=1
    
```
